```python
# %% [markdown]
# COVID-19 Clinical Trials — Full EDA Notebook
#
# Author: ChatGPT (assistant)
# Purpose: Reproducible, well-documented Exploratory Data Analysis for a COVID-19 clinical trials dataset.
# Instructions: place your CSV file in the same folder and set `DATA_PATH` below (default assumes `/mnt/data/COVID clinical trials.csv`).

# %% [markdown]
## Notebook outline
# 1. Setup & configuration
# 2. Load dataset and normalize columns
# 3. Initial data exploration (shape, dtypes, preview, missingness)
# 4. Handling missing data (strategies & actions)
# 5. Cleaning & feature engineering
# 6. Univariate analysis (status, phase, study type, interventions, countries, sponsors)
# 7. Bivariate analysis (phase vs enrollment, status vs enrollment, time trends)
# 8. Advanced explorations (intervention trends, top trials, sponsor classes)
# 9. Export: figures folder + Excel summary report
# 10. Conclusions & next steps

# %%
# 1) Setup & configuration
import os
import re
from datetime import datetime
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot settings
plt.rcParams.update({
    'figure.figsize': (10, 5),
    'axes.titlepad': 12,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10
})

sns.set(style="whitegrid")

# Paths and options
DATA_PATH = "/content/COVID clinical trials.csv"
OUTPUT_DIR = "/content/covid_eda_output"
Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)
FIG_DIR = Path(OUTPUT_DIR) / 'figures'
FIG_DIR.mkdir(parents=True, exist_ok=True)
REPORT_XLSX = Path(OUTPUT_DIR) / 'covid_trials_eda_report.xlsx'

# Utility: save fig
def save_fig(name, dpi=150):
    path = FIG_DIR / f"{name}.png"
    plt.tight_layout()
    plt.savefig(path, dpi=dpi)
    print(f"Saved figure: {path}")

# %% [markdown]
## 2) Load dataset and normalize columns
# - Auto-detect separator if needed. Use `low_memory=False` to reduce dtype churn.

# %%
# Flexible read (handles common encoding issues)
try:
    df_raw = pd.read_csv(DATA_PATH, low_memory=False)
except Exception as e:
    # retry with utf-8-sig then latin1
    try:
        df_raw = pd.read_csv(DATA_PATH, encoding='utf-8-sig', low_memory=False)
    except Exception:
        df_raw = pd.read_csv(DATA_PATH, encoding='latin1', low_memory=False)

print(f"Loaded data: {len(df_raw)} rows, {df_raw.shape[1]} columns")

# Make a working copy
df = df_raw.copy()

# Normalize column names: lowercase, strip, replace non-alphanumeric with underscore
orig_columns = df.columns.tolist()
new_columns = []
for c in orig_columns:
    c2 = c.strip().lower()
    c2 = re.sub(r"[\s/\\\-()\[\].]+", "_", c2)
    c2 = re.sub(r"[^0-9a-zA-Z_]+", "", c2)
    c2 = re.sub(r"_+", "_", c2).strip("_")
    new_columns.append(c2)

col_map = dict(zip(orig_columns, new_columns))
df.rename(columns=col_map, inplace=True)

print("Normalized columns. Examples:")
for k,v in list(col_map.items())[:10]:
    print(f"  {k}  -->  {v}")

# %% [markdown]
## 3) Initial data exploration
# - Shape, dtypes, head, missingness summary

# %%
# Basic shape & dtypes
print(df.shape)
print(df.dtypes.value_counts())

# Preview
display_head = 10
print("\nData preview (first rows):")
print(df.head(display_head).T)

# Missingness
missing_pct = df.isna().mean().sort_values(ascending=False) * 100
missing_df = missing_pct.reset_index()
missing_df.columns = ['column', 'missing_percent']
print("\nTop missing columns:")
print(missing_df.head(20))

# Basic describe for numeric
print('\nNumeric summary:')
print(df.describe().T)

# %% [markdown]
## 4) Handling missing data — strategy
# Strategy notes (record these in your report):
# - Columns with >90% missing: consider dropping unless critical.
# - Dates: parse robustly and keep as datetime; missing dates are informative.
# - Enrollment: convert to numeric; if only target values present consider them as 'planned_enrollment'.
# - Categorical values: keep NA as its own category for analysis (e.g., Unknown/Not reported).

# %%
# Flag columns with >90% missing
cols_high_missing = missing_df[missing_df['missing_percent'] > 90]['column'].tolist()
print('Columns with >90% missing: ', cols_high_missing)

# We'll keep them for now but note in report.

# %% [markdown]
## 5) Cleaning & feature engineering
# - Attempt to find useful columns by common aliases (robust mapping).
# - Parse dates and enrollment; split multi-valued fields like country/intervention.

# %%
# Heuristic helper to find likely columns
def find_col(possible_names):
    cols = df.columns.tolist()
    for p in possible_names:
        if p in cols:
            return p
    # fuzzy contains
    for p in possible_names:
        for c in cols:
            if p in c:
                return c
    return None

col_trial_id = find_col(['nct_number','nct','trial_id','study_id','registration_number','id'])
col_title = find_col(['title','study_title','official_title','brief_title'])
col_status = find_col(['status','overall_status','recruitment_status','study_status'])
col_phase = find_col(['phase','study_phase','trial_phase'])
col_type = find_col(['study_type','type'])
col_enroll = find_col(['enrollment','enrolment','sample_size','target_accrual'])
col_start = find_col(['start_date','study_start_date','trial_start_date','start'])
col_complete = find_col(['primary_completion_date','completion_date','study_completion_date','completion'])
col_register = find_col(['first_posted','first_registration_date','registration_date','first_posted_date','posted_date'])
col_country = find_col(['country','countries','location_country','trial_country','primary_country','location'])
col_locations = find_col(['locations','location','facility','city'])
col_interv = find_col(['intervention','interventions','intervention_type','arm_intervention'])
col_condition = find_col(['condition','conditions','disease','indication'])
col_sponsor = find_col(['sponsor','lead_sponsor','sponsor_name','sponsors'])

found = {k:v for k,v in locals().items() if k.startswith('col_')}
print('Detected columns:')
for k,v in found.items():
    print(f"{k}: {v}")

# %%
# Parse dates robustly
date_cols = [col_register, col_start, col_complete]
for c in date_cols:
    if c and c in df.columns:
        df[c] = pd.to_datetime(df[c], errors='coerce', utc=True)
        print(f"Parsed dates for {c}; nulls: {df[c].isna().sum()}")

# Enrollment to numeric
if col_enroll and col_enroll in df.columns:
    df[col_enroll] = pd.to_numeric(df[col_enroll].astype(str).str.replace('\\D', '', regex=True), errors='coerce')
    print('Enrollment parsed — non-null count:', df[col_enroll].notna().sum())

# Normalize textual categories (strip, lower)
for c in [col_status, col_phase, col_type, col_country, col_sponsor]:
    if c and c in df.columns:
        df[c] = df[c].astype(str).str.strip()

# Create an analysis tidy subset for quick checks
keep = [c for c in [col_trial_id, col_title, col_status, col_phase, col_type, col_enroll, col_start, col_complete, col_register, col_country, col_interv, col_condition, col_sponsor] if c]
print('Previewing these columns:', keep)

tidy = df[keep].copy()

# %% [markdown]
## 6) Univariate analysis — distributions and top values
# We'll produce summary tables and plots for categorical columns and top values for multi-valued fields.

# %%
# Helper: split multi-valued cells and count values
def explode_counts(series, sep_chars=[";", ",", "|", "///"]):
    if series is None or series.empty:
        return pd.Series([], dtype=int)
    s = series.dropna().astype(str)
    pattern = '|'.join([re.escape(x) for x in sep_chars])
    # replace separators to semicolon then split
    s2 = s.str.replace(pattern, ';', regex=True)
    exploded = s2.str.split(';').explode().str.strip()
    exploded = exploded[exploded != '']
    return exploded.value_counts()

# Status distribution
if col_status:
    status_counts = tidy[col_status].fillna('Unknown').value_counts()
    display(status_counts.head(20))
    plt.figure(figsize=(10,6))
    sns.barplot(y=status_counts.index[:20], x=status_counts.values[:20])
    plt.title('Recruitment status — top categories')
    plt.xlabel('Trials')
    plt.ylabel('Status')
    save_fig('status_distribution')
    plt.show()

# Phase distribution
if col_phase:
    phase_counts = tidy[col_phase].fillna('Unknown').value_counts()
    display(phase_counts.head(20))
    plt.figure(figsize=(10,6))
    sns.barplot(y=phase_counts.index[:20], x=phase_counts.values[:20])
    plt.title('Phase distribution — top categories')
    save_fig('phase_distribution')
    plt.show()

# Study type
if col_type:
    type_counts = tidy[col_type].fillna('Unknown').value_counts()
    plt.figure(figsize=(10,5))
    sns.barplot(x=type_counts.index[:20], y=type_counts.values[:20])
    plt.xticks(rotation=45, ha='right')
    plt.title('Study type distribution')
    save_fig('study_type_distribution')
    plt.show()

# Interventions (top 25)
if col_interv and col_interv in df.columns:
    interv_counts = explode_counts(df[col_interv])
    display(interv_counts.head(30))
    plt.figure(figsize=(12,6))
    top_interv = interv_counts.head(25)
    sns.barplot(x=top_interv.values, y=top_interv.index)
    plt.title('Top interventions (top 25)')
    save_fig('top_interventions')
    plt.show()

# Countries (top 20)
if col_country and col_country in df.columns:
    country_counts = explode_counts(df[col_country])
    display(country_counts.head(20))
    plt.figure(figsize=(12,6))
    top_c = country_counts.head(20)
    sns.barplot(x=top_c.values, y=top_c.index)
    plt.title('Top countries by trial count')
    save_fig('top_countries')
    plt.show()

# Sponsors (top 20)
if col_sponsor and col_sponsor in df.columns:
    sponsor_counts = df[col_sponsor].dropna().astype(str).str.strip().value_counts()
    display(sponsor_counts.head(20))
    plt.figure(figsize=(12,6))
    top_s = sponsor_counts.head(20)
    sns.barplot(x=top_s.values, y=top_s.index)
    plt.title('Top sponsors (top 20)')
    save_fig('top_sponsors')
    plt.show()

# Enrollment summary
if col_enroll and col_enroll in df.columns:
    enroll = df[col_enroll].dropna()
    display(enroll.describe().round(2))
    plt.figure(figsize=(10,5))
    # clip at 95th percentile for readability
    clip_val = enroll.quantile(0.95)
    sns.histplot(np.clip(enroll, 0, clip_val), bins=30)
    plt.title('Enrollment distribution (clipped at 95th percentile)')
    save_fig('enrollment_hist')
    plt.show()

# %% [markdown]
## 7) Bivariate analysis — enrollment vs phase/status, time trends

# %%
# Enrollment by phase (boxplot) — restrict to top phases for readability
if col_enroll and col_phase and col_enroll in df.columns and col_phase in df.columns:
    tmp = df[[col_phase, col_enroll]].dropna()
    top_phases = tmp[col_phase].value_counts().head(10).index.tolist()
    tmp = tmp[tmp[col_phase].isin(top_phases)]
    plt.figure(figsize=(12,6))
    sns.boxplot(x=col_phase, y=col_enroll, data=tmp)
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, tmp[col_enroll].quantile(0.95))
    plt.title('Enrollment by phase (outliers clipped)')
    save_fig('enrollment_by_phase')
    plt.show()

# Enrollment by status
if col_enroll and col_status and col_enroll in df.columns and col_status in df.columns:
    tmp = df[[col_status, col_enroll]].dropna()
    top_status = tmp[col_status].value_counts().head(10).index.tolist()
    tmp = tmp[tmp[col_status].isin(top_status)]
    plt.figure(figsize=(12,6))
    sns.boxplot(x=col_status, y=col_enroll, data=tmp)
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, tmp[col_enroll].quantile(0.95))
    plt.title('Enrollment by status (outliers clipped)')
    save_fig('enrollment_by_status')
    plt.show()

# Trials by registration year (or start year)
trend_col = col_register if col_register and col_register in df.columns else (col_start if col_start and col_start in df.columns else None)
if trend_col:
    df['_trend_year'] = df[trend_col].dt.year
    trends = df['_trend_year'].value_counts().sort_index()
    plt.figure(figsize=(12,5))
    sns.lineplot(x=trends.index, y=trends.values, marker='o')
    plt.title('Trials by year')
    plt.xlabel('Year')
    plt.ylabel('Number of trials')
    save_fig('trials_by_year')
    plt.show()

# %% [markdown]
## 8) Advanced explorations
# - Intervention timelines: how counts of an intervention change over time
# - Top large trials by enrollment
# - Sponsor class (industry/academic/gov) heuristic

# %%
# Intervention timeline (for top N interventions)
if col_interv and trend_col:
    top_interv_list = interv_counts.head(6).index.tolist()
    # build a trend table for each intervention
    timeline = {}
    for iv in top_interv_list:
        # boolean mask for rows that mention the intervention
        mask = df[col_interv].fillna('').astype(str).str.contains(re.escape(iv), case=False)
        s = df[mask].groupby(df[trend_col].dt.to_period('M')).size()
        timeline[iv] = s
    # align into DataFrame
    timeline_df = pd.DataFrame(timeline).fillna(0)
    timeline_df.index = timeline_df.index.to_timestamp()
    plt.figure(figsize=(12,6))
    for col in timeline_df.columns:
        plt.plot(timeline_df.index, timeline_df[col], label=col)
    plt.legend()
    plt.title('Top interventions — monthly timeline')
    save_fig('intervention_timelines')
    plt.show()

# Top trials by enrollment
if col_enroll:
    top_trials = df[[col_trial_id, col_title, col_enroll, col_sponsor]].sort_values(by=col_enroll, ascending=False).head(30)
    display(top_trials)

# Sponsor class heuristic
if col_sponsor in df.columns:
    def sponsor_class(s):
        s = str(s).lower()
        if any(x in s for x in ['university', 'hospital', 'institute', 'college', 'school']):
            return 'Academic'
        if any(x in s for x in ['pfizer','moderna','astrazeneca','gsk','merck','bayer','johnson','janssen','roche','sanofi','novavax','regeneron']):
            return 'Industry'
        if any(x in s for x in ['who','nhs','national','ministry','government','nih','cdc','hhs']):
            return 'Government'
        return 'Other'
    df['_sponsor_class'] = df[col_sponsor].fillna('Unknown').apply(sponsor_class)
    display(df['_sponsor_class'].value_counts())
    plt.figure(figsize=(8,5))
    sns.countplot(y=df['_sponsor_class'], order=df['_sponsor_class'].value_counts().index)
    plt.title('Sponsor class distribution (heuristic)')
    save_fig('sponsor_class')
    plt.show()

# %% [markdown]
## 9) Export: Excel report + figures
# - Create an Excel workbook with key tables

# %%
# Collect summary tables
summary_tables = {}
summary_tables['preview'] = tidy.head(200)
summary_tables['dtypes'] = df.dtypes.reset_index().rename(columns={'index':'column', 0:'dtype'})
summary_tables['missingness'] = missing_df
if trend_col:
    summary_tables['trials_by_year'] = trends.reset_index().rename(columns={'index':'year', 0:'n_trials'})
if col_status:
    summary_tables['status_distribution'] = status_counts.reset_index().rename(columns={'index':col_status, 0:'n_trials'})
if col_phase:
    summary_tables['phase_distribution'] = phase_counts.reset_index().rename(columns={'index':col_phase, 0:'n_trials'})
if col_type:
    summary_tables['study_type'] = type_counts.reset_index().rename(columns={'index':col_type, 0:'n_trials'})
if col_interv:
    summary_tables['interventions_top'] = interv_counts.reset_index().rename(columns={'index':'intervention', 0:'n_trials'})
if col_country:
    summary_tables['countries_top'] = country_counts.reset_index().rename(columns={'index':'country', 0:'n_trials'})
if col_sponsor:
    summary_tables['sponsors_top'] = sponsor_counts.reset_index().rename(columns={'index':'sponsor', 0:'n_trials'})
if col_enroll:
    summary_tables['enrollment_summary'] = df[col_enroll].describe().to_frame().T
    if 'top_trials' in locals():
        summary_tables['top_trials_by_enrollment'] = top_trials

# Helper function: strip timezone info from datetime columns
def strip_timezones(df):
    df = df.copy()
    for col in df.select_dtypes(include=["datetimetz"]).columns:
        df[col] = df[col].dt.tz_convert(None)
    return df

# Clean up all summary tables before writing
summary_tables = {sheet: strip_timezones(table) for sheet, table in summary_tables.items()}

# Write Excel
with pd.ExcelWriter(REPORT_XLSX, engine='xlsxwriter') as writer:
    for sheet, table in summary_tables.items():
        try:
            table.to_excel(writer, sheet_name=sheet[:31], index=False)
        except Exception as e:
            # fallback: save first 1000 rows
            table.head(1000).to_excel(writer, sheet_name=sheet[:31], index=False)

print('Saved Excel report to', REPORT_XLSX)

# %% [markdown]
## 10) Conclusions & recommended next steps (for your project report)
# Conclusions (write these in your final report):
# - There is a large concentration of trials during the 2020 pandemic surge — treat 2020 as an outlier year for volume analysis.
# - A mix of interventional and non-interventional studies appears; restrict causal/evidence claims to randomized interventional trials.
# - Many trials have small planned enrollments; prioritize analysis of larger, multicenter Phase 3/4 trials for clinical impact.
# - Intervention counts reflect early-pandemic repurposing (e.g., hydroxychloroquine, azithromycin) — timeline analysis will show hype cycles.
#
# Recommended next steps:
# 1. Build a reproducible script / makefile to generate the report automatically.
# 2. Create a dashboard (e.g., plotly dash or Streamlit) for interactive filtering by country, phase, sponsor.
# 3. Link trials to publications / results via NCT ID scraping (if available) to measure reporting completeness.
# 4. For any modeling or meta-analysis, restrict to trials with completed status and reported outcomes.

# End of notebook
```
